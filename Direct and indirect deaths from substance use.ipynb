## Visualization

This inolves understanding my datasets by placing it in a visual context so that patterns, trends and correlations that might not otherwise be detected can be exposed.

```python
#rows representing direct and indirect death from the use of drugs for the world
df4= df1.loc[df1['Entity'] == 'World']
```

```python
df4
```

```python
df4.sum()
```

```python
# The Bar chart shows the the number of people who have died from direct and indirect usage of drugs from  1990 to 2017 in the world
```

```python
df4[['Tobacco (risk factor) (deaths)','Alcohol (risk factor) (deaths)', 'Illicit drugs (risk factor) (deaths)', 'Illicit drug use disorders (deaths)', 'Alcohol use disorders (deaths)']].sum().plot.bar(figsize=(20, 10), title = 'Figure 1: Direct and indirect deaths cause by drugs from 1990 to 2017' )
plt.show()
```

Indirect deaths(represents first three bar) which results from substance use acting as a risk factor for the development of various diseases and injury 
direct deaths from substance use disorders (the last two bar). These are deaths which result from alcohol or illicit
drug use overdoses.
–

From my bar chart  over 285million people have been killed indirectly from the use of tobacco, alcohol and illicit drugs from 1990 to 2017. And over 7.7million people from direct usage

```python
#rows representing direct and indirect death from the use of drugs for the year 2017
df5= df4.loc[df1['Year'] == 2017]
df5
```

```python
df5[['Tobacco (risk factor) (deaths)','Alcohol (risk factor) (deaths)', 'Illicit drugs (risk factor) (deaths)', 'Illicit drug use disorders (deaths)', 'Alcohol use disorders (deaths)']].sum().plot.bar(figsize=(16, 8), color= 'green', title = 'Figure 2: Direct and indirect deaths from the use of drugs for 2017' )
plt.show()
```

In the visualization I summarize the results of the Global Burden of Disease study on the number of deaths attributed to substance use: this is differentiated between indirect deaths as a risk factor from smoking, alcohol and illicit drug use (represents first three bar) and direct deaths from substance use disorders of alcohol and illicit drugs (irepresents the last two bar).
According to our graph substance use was responsible for over 11.8 million deaths in 2017. This is one in five deaths globally
We see that premature deaths from smoking is the large contributor to this. Overall, premature deaths from substance use greatly outnumber direct deaths from overdoses.
